# 1.3 Apply hierarchical clustering to scaled data 

```python
# Hierarchical Agglomerative Clustering

#scipy

#sklearn

from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
```

```python
Z = linkage(data_scaled,method='ward')
```

```python
Z
```

```python
dend = dendrogram(Z)

plt.axhline(10)
```

```python
dend = dendrogram(Z,
                 truncate_mode='lastp',
                 p = 25,# we are looking at the last 25 merges
                 )
```

```python
#Method1 - 3 clusters as final output
cluster_labels = fcluster(Z,3,criterion='maxclust')
```

```python
cluster_labels
```

```python
df['clusters'] = cluster_labels
```

```python
df_clust = df.groupby('clusters').mean()
```

# K Means Clustring Method

```python
# Choosing Optimal no.of Clusters

# WSS Plot

K_means = KMeans(n_clusters=3,random_state=1)
```

```python
clust_3 = K_means.fit_predict(data_scaled)
```

```python
clust_3
```

```python
K_means.inertia_       # Total WSS
```

```python
wss = []

for i in range(1,11):
  KM = KMeans(n_clusters=i,random_state=1)
  KM.fit(data_scaled)
  wss.append(KM.inertia_)
```

```python
wss
```

```python
plt.plot(range(1,11),wss)
```
